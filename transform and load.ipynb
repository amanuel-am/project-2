```python
# import libraries
import pandas as pd
```

```python
# load book data
book_df = pd.read_csv('books_toscrape.csv')
```

```python
book_df
```

```python
# load book details
book_detail_df = pd.read_csv('books_detail.csv')
```

```python
book_detail_df
```

```python
# Truncate descriptions longer than 1000 characters
book_detail_df['Description'] = book_detail_df['Description'].apply(lambda x: x if len(x) <= 1000 else x[:1000])
# Remove both single and double quotation marks
book_detail_df['Description'] = book_detail_df['Description'].str.replace('"', '').str.replace("'", "")
```

```python
book_detail_df.dtypes
```

```python
# change rating column into stars
print(book_df['Rating'].value_counts())
rating_to_stars = {
    'One': '★☆☆☆☆',
    'Two': '★★☆☆☆',
    'Three': '★★★☆☆',
    'Four': '★★★★☆',
    'Five': '★★★★★'
}
```

```python
# Apply the mapping to the 'Rating' column
book_df['Rating'] = book_df['Rating'].map(rating_to_stars)
```

```python
book_df
```

```python
# Check for duplicate values
duplicate_book_rows = book_df[book_df.duplicated(keep=False)]
duplicate_book_rows
```

```python
duplicate_detail_rows = book_detail_df[book_detail_df.duplicated(keep=False)]
duplicate_detail_rows
```

```python
book_df.iloc[book_df.index[book_df['Title'].str.contains('The Star-Touched Queen', case=False, na=False)].tolist()]
```

```python
# fill empty values
book_df = book_df.fillna ('N/A')
book_detail_df = book_detail_df.fillna('N/A')
```

```python
book_df.dtypes
```

```python
#change price column from GBP to USD
book_df['Price'] = book_df['Price'].replace({'Â£': ''}, regex=True)
book_df['Price']= book_df['Price'].astype(float)
book_df['Price'] = book_df['Price']*1.28
book_df['Price'] = book_df['Price'].apply(lambda x: f'${x:.2f}')
```

```python
book_df
```

```python
# Save  cleaned DataFrame to a new CSV file
book_df.to_csv('cleaned_book.csv', index=False)
book_detail_df.to_csv('cleaned_books_detail.csv', index=False)
```
